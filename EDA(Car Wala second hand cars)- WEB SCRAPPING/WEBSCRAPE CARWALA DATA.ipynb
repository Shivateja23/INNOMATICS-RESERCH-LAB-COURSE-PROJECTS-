```python
from bs4 import BeautifulSoup
# Assume `html` contains the rendered HTML from Selenium (or another method)
soup = BeautifulSoup(html, "lxml")

# Find all span tags that contain the mileage, fuel type, and location info.
info_spans = soup.find_all("span", class_="o-eemiLE o-zmksK")
print("Number of info spans found:", len(info_spans))

# Initialize a list to hold data for each car
car_data = []

# Process each span individually
for span in info_spans:
    text = span.get_text(strip=True)
    # Split the string by "|" – expecting three parts.
    parts = [part.strip() for part in text.split("|")]
    if len(parts) >= 3:
        mileage, fuel, location = parts[0], parts[1], parts[2]
        car_data.append({
            "mileage": mileage,
            "fuel": fuel,
            "location": location
        })
    else:
        # If the format is unexpected, print it out for debugging.
        print("Unexpected format:", text)

# Now, print the extracted information for each car
for car in car_data:
    print("Mileage:", car["mileage"])
    print("Fuel Type:", car["fuel"])
    print("Location:", car["location"])
    print("-" * 40)
```

```python
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup

def parse_info(text):
    """
    Process an info span's text.
    The expected format is either:
      "value1|value2|value3"
    or if two metrics are present, a delimiter like " | KM Driven:" separates them:
      "value1a|value2a|value3a | KM Driven: value1b|value2b|value3b"
    Returns a dictionary with keys:
      If two parts are present, keys:
         metric_1_mileage, metric_1_fuel, metric_1_location,
         metric_2_mileage, metric_2_fuel, metric_2_location
      Else:
         mileage, fuel, location
    """
    data = {}
    # Check if the delimiter exists (case-insensitive search)
    delimiter = " | KM Driven:"
    if delimiter in text:
        parts = text.split(delimiter)
        part1 = parts[0].strip()
        part2 = parts[1].strip()
        fields1 = [x.strip() for x in part1.split("|")]
        fields2 = [x.strip() for x in part2.split("|")]
        if len(fields1) >= 3:
            data["metric_1_mileage"]   = fields1[0]
            data["metric_1_fuel"]      = fields1[1]
            data["metric_1_location"]  = fields1[2]
        if len(fields2) >= 3:
            data["metric_2_mileage"]   = fields2[0]
            data["metric_2_fuel"]      = fields2[1]
            data["metric_2_location"]  = fields2[2]
    else:
        fields = [x.strip() for x in text.split("|")]
        if len(fields) >= 3:
            data["mileage"]  = fields[0]
            data["fuel"]     = fields[1]
            data["location"] = fields[2]
    return data

def scrape_page(page, driver):
    """
    Loads the given page (page number) and returns a list of dictionaries,
    one for each car listing, containing title, price, EMI and additional info.
    """
    # Construct URL – for page 1 you might not need "page-1", but here we use it for consistency.
    url = f"https://www.carwale.com/used/hyderabad/page-{page}/"
    print(f"Scraping page {page}: {url}")
    driver.get(url)
    time.sleep(8)  # Wait for dynamic content to load

    html = driver.page_source
    soup = BeautifulSoup(html, "lxml")
    
    # Extract elements using the provided selectors
    titles = soup.find_all("h3", class_="o-bqHweY ERtQDA o-bfyaNx o-bNxxEB o-cJrNdO o-byFsZJ o-fzoTzs o-bkmzIL")
    info_spans = soup.find_all("span", class_="o-eemiLE o-zmksK")
    price_elements = soup.find_all("span", class_="o-Hyyko o-cJrNdO o-eqqVmt o-fzpibr")
    emi_elements = soup.find_all("div", class_="o-cwUopP o-eemiLE o-fnsUKv o-fzoTpF")
    
    # Debug: print counts
    print("  Titles found:", len(titles))
    print("  Info spans found:", len(info_spans))
    print("  Price elements found:", len(price_elements))
    print("  EMI elements found:", len(emi_elements))
    
    listings = []
    
    # We assume that the primary listing elements (titles, prices, EMI) are aligned.
    # Their count should be equal (e.g. 32 listings per page).
    num_listings = len(titles)
    for i in range(num_listings):
        record = {}
        record["title"] = titles[i].get_text(strip=True) if i < len(titles) else None
        record["price"] = price_elements[i].get_text(strip=True) if i < len(price_elements) else None
        record["emi"] = emi_elements[i].get_text(strip=True) if i < len(emi_elements) else None
        
        # Process the info span for this listing.
        # (Assuming the i-th info span corresponds to the i-th listing.)
        if i < len(info_spans):
            info_text = info_spans[i].get_text(strip=True)
            info_data = parse_info(info_text)
            record.update(info_data)
        listings.append(record)
    
    return listings

def main():
    total_pages = 240  # Total pages to scrape
    all_listings = []
    
    # Set up Selenium with headless Chrome
    options = Options()
    options.add_argument("--headless")
    options.add_argument("--disable-gpu")
    options.add_argument("--window-size=1920,1200")
    driver = webdriver.Chrome(options=options)
    
    for page in range(1, total_pages + 1):
        try:
            page_listings = scrape_page(page, driver)
            print(f"  Scraped {len(page_listings)} listings from page {page}")
            all_listings.extend(page_listings)
        except Exception as e:
            print(f"Error on page {page}: {e}")
    
    driver.quit()
    
    # Convert the aggregated data into a DataFrame
    df = pd.DataFrame(all_listings)
    print("\nTotal listings scraped:", len(df))
    # Optionally, save the DataFrame to a CSV file
    df.to_csv("carwale_used_cars.csv", index=False)
    print("Data saved to 'carwale_used_cars.csv'")
    
if __name__ == "__main__":
    main()
```
